## Uso de los filtros

### Filtro 1

```python
filro_0a = df.iloc[: , [0,11]]
filro_0a
```

```python
#a) Los registros de 5 "host name" diferentes
filtroA = df[(df["host_name"] == "Wakana") | (df["host_name"] == "Kei") | (df["host_name"] == "Jin") | (df["host_name"] == "Rao") | (df["host_name"] == "Yukiko")]
#(["Wakana", "Kei", "Jin", "Rao", "Yukiko"])]
filtroA
```

## Filtro 2

```python
#b) Los registros de los host que se hallan unido a Airbnb "antes del año 2019"
filtroB = df[df["host_since"] < "2019-01-01"]
filtroB
```

### Filtro 3

```python
#c) Los registros de los host que responden "a mas tardar en 1 día"
filtroC = df[df["host_response_time"] == "within a day"]
filtroC
```

### Filtro 4

```python
#d) Los registros que tengan "al menos 5 accommodates"
filtroD = df[df["accommodates"] >= 5]
filtroD
```

### Filtro 5

```python
#e) Los registros de los tipos de cuarto: "Entire home" y "Private room"
filtroE = df[(df["room_type"] == "Entire home/apt") | (df["room_type"] == "Private room")]
filtroE
```

### Filtro 6

```python
#f) Los registros que "a lo mucho cuenten con 4 camas"
filtroF = df[df["beds"] <= 4]
filtroF
```

### Filtro 7

```python
#g) Los registros que "tengan 1 baño o que sean superhost" pero con la condicion
# de que "todos sean instant_bookable"
filtroG = df[((df["bathrooms"] == 1) | (df["host_is_superhost"] == "t")) & (df["instant_bookable"] == "t")]
filtroG
```

### Filtro 8

```python
#h) Los registros cuyo "reviw_scores_cleanliness sea menor de 3 y reviw_scores_rating sea mayor de 3"
filtroH = df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)]
filtroH
```
