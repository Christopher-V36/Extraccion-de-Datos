```python
#Librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Carga del archivo

```python
#Carga desde un archivo .xlsx sin indice
df = pd.read_excel("Detalle precios y productos fabricados 2022.xlsx")
df
```

```python

#Verificamos información del DataFrame
df.info()
```

## Uso de los filtros

### Filtro 1

```python

#Filtro por objeto
filtro1 = df [(df ["NOMBRE_VENDEDOR"] == "AARON NUÑO CORDOVA")]
filtro1
```

### Filtro 2

```python
#Filtro por filas
filtro2 = df.iloc[0:6, :]      #filas consecutivas
#filtro2 = df.iloc[[2, 5], :]  #filas no consecutivas
filtro2
```

### Filtro 3

```python
#Filtro por columnas
#filtro3 = df.iloc[: , 1:3]        #columnas consecutivas
filtro3 = df.iloc[: , [1, 2, 11]]   #columnas no consecutivas
filtro3
```

### Filtro 4 

```python
#Carga de archivo .xlsx con indice 
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=2)
df1
```

```python

#Filtro filas vs columnas
filtro4 = df1.loc[["LETICIA RAMIREZ HERNANDEZ", "AARON NUÑO CORDOVA"], ["PRECIO_UNITARIO", "CANT"]]
filtro4
```

### Filtro 5

```python

#Filtro por cabecera, mostrará las primeras 10 filas extraidas de la base 
#de datos (cuando el parentesis esté vacío muestra las primeras 5)
filtro5 = df.head(3)
filtro5
```

### Filtro 6

```python
#Filtro por cola, mostrará las últimas 10 filas extraidas de la base 
# de datos (cuando el parentesis esté vacío muestra  las últimas 5)
filtro6 = df.tail(3)
filtro6
```

### Filtro 7

```python
#Filtro por comparación, en este flitro se selecciona una variable 
# de la que se requiera extraer datos con uns criterio especificado. 
filtro7 = df[df["PRECIO_UNITARIO"] > 25000]
filtro7
```

### Filtro 8

```python
#Filtro Y, se trabaja con el conector lógico & para extraer datos 
#específicos de la base de datos.
filtro8 = df[(df["PRECIO_UNITARIO"] > 6000) & (df["CANT"] > 10)]
filtro8
```

### Filtro 9

```python
#Filtro O, se trabaja con el conector lógico | para extraer datos 
filtro9 = df[(df["COSTO_UNITARIO"] > 10000) | (df["FECHA_DOC"] == "2022-09-07")]
filtro9
```

### Filtro 10

```python
#Filtro not, se trabaja con el conector lógico ~ para extraer datos 
filtro10 = df[~(df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL")]
filtro10
```

```python
filtro5.to_csv("Filtro por cabecera.csv")
#Filtro por cabecera, muestra las primeras 3 filas
```

```python
filtro7.to_csv("Filtro por comparación.csv")
#Filtro por comparación, muestra los datos con un precio unitario mayor a 25000
```

```python
filtro8.to_csv("Filtro Y.csv")
#Filtro Y, muestra los datos con un precio unitario mayor a 6000 y una cantidad mayor a 10
```
